```python
SN2020oi = pd.read_csv("/Users/alexgagliano/Documents/Research/2020oi/data/photometry/2020oiphotometry_clean_wUVOT.csv")
SN2020oi = SN2020oi[SN2020oi['FLT'] != 'orange']
SN2020oi.loc[SN2020oi['FLT'] == 'up', 'FLT'] = 'u'
SN2020oi.loc[SN2020oi['FLT'] == 'gp', 'FLT'] = 'g'
SN2020oi.loc[SN2020oi['FLT'] == 'g-ZTF', 'FLT'] = 'g'
SN2020oi.loc[SN2020oi['FLT'] == 'rp', 'FLT'] = 'r'
SN2020oi.loc[SN2020oi['FLT'] == 'r-ZTF', 'FLT'] = 'r'
SN2020oi.loc[SN2020oi['FLT'] == 'ip', 'FLT'] = 'i'

#Spectral radiance in units of erg/s/Angstrom
#1 Jy = 1.e-23 erg/s/cm^2/Hz
#Fnu = Flambda x (lambda**2) / c
sns.set_context("poster")

for date in allFiles:
    c = 2.998e10 #cm/s
    dm = 31.16
    dist = 1e-5*3.086e24
    
    real_data = pd.read_csv("%i_Data.csv"%date)
    model_data = pd.read_csv("%i_Model.csv"%date)

    #plt.plot(real_data['wave'],real_data['FluxData'], 'o')
    model_phase = model_data['ph'].values[0]
    print(model_phase)
    SN2020oi_cut = SN2020oi[np.abs(SN2020oi['MJD'] - model_phase) < 0.2]
    SN2020oi_cut['wle'] = np.array([wle[x] for x in SN2020oi_cut['FLT']])
    SN2020oi_cut['ext_corr'] = np.array([extco[x] for x in SN2020oi_cut['FLT']])
    
    # extinction-correction
    ebv = 0.174+0.028
    ebv2 = 0.174-0.028
    #ebv2 = 0.153 #horesh's value
    # Subtract foreground extinction using input E(B-V) and coefficients from YES
    SN2020oi_cut['MAG'] -= SN2020oi_cut['ext_corr']*ebv
    print(SN2020oi_cut['FLT'].values)
    print(SN2020oi_cut['ext_corr'].values*ebv - SN2020oi_cut['ext_corr'].values*ebv2)
    
    SN2020oi_cut['AbsMag'] = SN2020oi_cut['MAG']-dm
    
    #calculate zeropoints
    fref = np.array([zp[x]*1e-11 for x in SN2020oi_cut['FLT']])    
    SN2020oi_cut['flux'] = 4*np.pi*dist**2*fref*10**(-0.4*SN2020oi_cut['AbsMag'])
    #this gives flux in erg/s/Angstrom
    #first, get rid of the distance to get to erg/s/cm^2/Angstrom
    SN2020oi_cut['flux'] /= 4*np.pi*dist**2
    
    #next, use Fnu = Flambda x (lambda**2) / c
    SN2020oi_cut['flux'] *= SN2020oi_cut['wle']
#    SN2020oi_cut['flux'] *= SN2020oi_cut['wle']**2/c
#    SN2020oi_cut['flux'] /= 1.e-23
#    #convert Fnu to nu Fnu
    
    
    # convert mag errors to flux errors
    #ferr = 2.5/np.log(10) * flux * errs
    
    #First, get rid of the distance to get to erg/s/cm^2/Angstrom 
    model_data['BBody_Fit'] /= 4*np.pi*dist**2
    #next, use Fnu = Flambda x (lambda**2) / c
    model_data['BBody_Fit'] *= model_data['wave']
 #   model_data['BBody_Fit'] *= model_data['wave']**2/c
 #   #next, convert to Janskies
 #   model_data['BBody_Fit'] /= 1.e-23
    
    #convert Fnu to nu Fnu
    

    Tbb = BBody_temps.loc[np.abs(BBody_temps['ph'] - model_phase) < 0.01, 'T_bb'].values[0]
    Rbb = BBody_temps.loc[np.abs(BBody_temps['ph'] - model_phase) < 0.01, 'R_bb'].values[0]
    
    uvot_done = 0
    if (len(SN2020oi_cut) > 1) and (uvot_done == 0):
        plt.figure(figsize=(10,7))
        plt.plot(model_data['wave'], model_data['BBody_Fit'],lw=4)
        plt.plot(SN2020oi_cut['wle'], SN2020oi_cut['flux'], 'o')
        for uv in np.array(['UVM1', 'UVW1', 'UVW2']):
            if uv in SN2020oi_cut['FLT']:
                uvot_pt = SN2020oi_cut[SN2020oi_cut['FLT'].isin(['UVM1', 'UVW1', 'UVW2'])]
                plt.plot(uvot_pt['wle'], uvot_pt['flux'], '*', ms=10)
                uvot_done = 1
        plt.xscale("log")
        plt.yscale("log")
        #plt.xlim((500, 10000))
        #plt.ylim((1.e1, 1.e20))
        plt.xlabel("Wavelength (AA)")
        plt.ylabel(r"$\nu F_\nu$ (erg/s/cm$^2$)")
        plt.ylim(ymin=1.e-5)
        
        plt.tick_params(which='major', axis='x', direction='in', bottom=True, top=True, length=16, width=2)
        plt.tick_params(which='major', axis='y', direction='in', left=True, right=True, length=16, width=2)
        plt.tick_params(which='minor', axis='x', direction='in', bottom=True, top=True, length=8, width=2)
        plt.minorticks_on()
        plt.tick_params(which='minor', axis='y', direction='in', left=True, right=True, length=8, width=2)
        plt.text(1.e2, 2.e4, "$t$ (MJD) = %.2f"%model_phase)
        plt.text(1.e2, 5.e2, r"$T_{BB}$ = %.2f K"%Tbb)
        plt.text(1.e2, 1.e1, r"$R_{BB}$ = %.2f x $10^{14}$ cm"%float(Rbb/1.e14)) 
#        plt.savefig("2020oi_BBodyFit_MJD%2f.png"%date,dpi=300, bbox_inches='tight')
      #  break
```
