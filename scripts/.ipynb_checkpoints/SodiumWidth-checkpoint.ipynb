```python
EW = equivalent_width(cont_norm_spec, regions=SpectralRegion(5912*u.AA, 5930*u.AA))
EBV =10**(1.17*EW.value - 1.85)
print("EW = ", EW)
print("EBV = ", EBV)
```

```python
# creating a loop here: 
center = 5921 #angstroms
rad = np.linspace(2, 300, num=500)
widths = rad*2
sr = SpectralRegion(589*u.um, 590*u.um)
sr2 = SpectralRegion(594*u.um, 596*u.um)
sr3 = SpectralRegion(591*u.um, 593*u.um)
EW_set = []
EBV_set = []
for width in widths:
    cutout = lris_spec[lris_spec['Wavelength(A)'] < center+width]
    cutout = cutout[cutout['Wavelength(A)'] > center-width]
    
    norm_wav = cutout['Wavelength(A)'].values*u.Angstrom#
    norm_flux = cutout['Flux'].values*u.Unit('erg cm-2 s-1 Angstrom-1') 
    spec= Spectrum1D(spectral_axis=norm_wav, flux=norm_flux) 
    cont_norm_spec = spec / fit_continuum(spec, exclude_regions=[sr, sr2, sr3])(spec.spectral_axis)
    EW = equivalent_width(cont_norm_spec, regions=SpectralRegion(5912*u.AA, 5930*u.AA))
    EBV =10**(1.17*EW.value - 1.85)
    EW_set.append(EW.value)
    EBV_set.append(EBV)
EW_set = np.array(EW_set)
EBV_set = np.array(EBV_set)
```
