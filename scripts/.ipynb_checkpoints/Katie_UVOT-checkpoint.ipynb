```python
import glob
import os 
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns 

UVM2 = []
UVW2 = []
files = glob.glob('../data/Ic_SwiftTemplate/*.dat')
for file in files:
    SN = file.split("/")[-1].split("_")[0]
    df = pd.read_csv(file, delim_whitespace=True)
    startTime = np.min(df['MJD[days]'])
    df['phase'] = df['MJD[days]'] - np.min(df['MJD[days]'])
    df['SN'] = SN
    df.dropna(subset=['Mag'], inplace=True)
    UVM2.append(df[df['Filter'] == 'UVM2'])
    UVW2.append(df[df['Filter'] == 'UVW2'])
df_UVM2 = pd.concat(UVM2, ignore_index=True)
df_UVW2 = pd.concat(UVW2, ignore_index=True)
```

```python
#plt.plot(df_UVM2['phase'], df_UVM2['Mag'], 'o')
dfUVM2_Ic = df_UVM2[df_UVM2['SN'] != 'SN2016coi']
sns.scatterplot(x="phase", y="Mag", hue="SN",data=dfUVM2_Ic)
plt.xlim((-1, 10))
plt.gca().invert_yaxis()
```

```python
#plt.plot(df_UVM2['phase'], df_UVM2['Mag'], 'o')
dfUVW2_Ic = df_UVW2[df_UVW2['SN'] != 'SN2016coi']
sns.scatterplot(x="phase", y="Mag", hue="SN",data=dfUVW2_Ic)
plt.xlim((-1, 10))
plt.gca().invert_yaxis()
```
