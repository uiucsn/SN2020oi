```python
#marginalize over mass as well:
Age = []
MargLike = []
MargProb = []

for age in np.unique(DF_margZ['Age']):
    Age.append(age)
    tempDF = DF_margZ[DF_margZ['Age']==age]
    #print(tempDF)
    MargLike.append(np.sum(tempDF['margLike']))
    MargProb.append(np.sum(tempDF['Prob']))
```

```python
from sklearn.metrics import auc
```

```python
Prospector_Ages = DF_ages['Age'].values
Prospector_Probs = DF_ages['MargProb']#/np.nansum(DF_ages['MargProb'])
BPASS_Ages = (10**BPASS_Model['Ages'])/1.e9
BPASS_Probs = BPASS_Model['BPASS_ProbDensity']#/np.nansum(BPASS_Model['BPASS_ProbDensity'])
import scipy.interpolate

auc_bpass = auc((10**BPASS_Model['Ages'])/1.e9, BPASS_Model['BPASS_ProbDensity'])
auc_DF = auc(DF_ages['Age'], DF_ages['MargProb'])
auc_starlight = auc(ages/1.e9, age_prob)

f_starlight = scipy.interpolate.interp1d(ages/1.e9, age_prob/auc_starlight)
f_prosp = scipy.interpolate.interp1d(Prospector_Ages, Prospector_Probs/auc_DF)

starlight_modelProbs = f_starlight(BPASS_Ages)
prosp_modelProbs = f_prosp(BPASS_Ages)

joint_prob = prosp_modelProbs*BPASS_Probs#*starlight_modelProbs

auc_joint = auc(BPASS_Ages, joint_prob)

joint_prob /= auc_joint
```

```python
#fitting it to a gaussian
from scipy.optimize import curve_fit

def gaussian(x, mean, amplitude, standard_deviation):
    return amplitude * np.exp( - (x - mean)**2 / (2*standard_deviation ** 2))
  
popt, _ = curve_fit(f=gaussian,xdata=BPASS_Ages,ydata=joint_prob, p0=[1., 0., 1.])

plt.plot(BPASS_Ages, joint_prob, c='k')
plt.fill_between(BPASS_Ages,0, joint_prob, alpha=0.4, color='k')
plt.plot(BPASS_Ages, gaussian(BPASS_Ages, *popt), label='fit')
```

```python
BPASS_Photometry_Single =  pd.read_csv("/Users/alexgagliano/Documents/Research/2020oi/data/derived_data/bestFit_singleBPASS_ExtUncertainty.csv")
BPASS_Photometry_Double = pd.read_csv("/Users/alexgagliano/Documents/Research/2020oi/data/derived_data/bestFit_binaryBPASS_ExtUncertainty.csv")
```

```python
BPASS_Photometry_Single['chi_sq_test'] = BPASS_Photometry_Single['chi_sq']/100.
BPASS_Photometry_Double['chi_sq_test'] = BPASS_Photometry_Double['chi_sq']/100.
```

```python
ages = np.unique(BPASS_Photometry_Single['logAge'])
Zs = np.unique(BPASS_Photometry_Single['Z'])
singleModel_probs = []
import numpy as np
import matplotlib.pyplot as plt
cols = sns.color_palette('Dark2', 9)

plt.figure(figsize=(10,7))
for i in np.arange(len(Zs)):
    tempBPASS = BPASS_Photometry_Single[BPASS_Photometry_Single['Z']== Zs[i]]
    P_photSingle = np.exp(-tempBPASS['chi_sq_test'].values/2.0)
    P_photSingle_norm = P_photSingle/np.sum(P_photSingle, axis=0)
  #  print(np.nanmax(P_photSingle))
    plt.plot(ages, P_photSingle_norm, '--',label="Z=%.2e"%Zs[i],c=cols[i])
    
    tempBPASS = BPASS_Photometry_Double[BPASS_Photometry_Double['Z']== Zs[i]]
    
    P_photDouble = np.exp(-tempBPASS['chi_sq_test'].values/2.0)
    P_photDouble_norm = P_photDouble/np.sum(P_photDouble, axis=0)
    #print(np.nanmax(P_photDouble))
    plt.plot(ages, P_photDouble_norm, c=cols[i])
plt.legend()
plt.xlabel("Age")
plt.xlim((6, 8))
plt.show()
```
