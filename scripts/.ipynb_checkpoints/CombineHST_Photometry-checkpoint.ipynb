```python
data_HST_old = pd.read_csv("/Users/alexgagliano/Documents/Research/2020oi/data/photometry/HST_oldData.csv", delimiter='&')
```

```python
data_HST_old['Filter'].values[0]
```

```python
newMag = []
newMagErr = []

for idx, row in data_HST_old.iterrows():
    tempDF = data_HST[np.abs(data_HST['MJD'] - row['MJD'])<0.1]
    tempDF = tempDF[tempDF['FILT']==row['Filter']]
    newMag.append(tempDF['MAG(AB)'].values[0])
    newMagErr.append(tempDF['MAGERR'].values[0])
```

```python
data_HST_old['Mag'] = newMag
data_HST_old['Magerr'] = newMagErr

data_HST_old.to_csv("/Users/alexgagliano/Documents/Research/2020oi/data/photometry/HST_PhysicalAperture.csv",index=False)
```
