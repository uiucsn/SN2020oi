```python
SN2013dx_min8 = pd.read_csv("/Users/alexgagliano/Documents/Research/2020oi/data/reference_data/SN2013dx_SP1.csv")
SN2002ap_min4 = pd.read_csv("/Users/alexgagliano/Documents/Research/2020oi/data/reference_data/SN2002ap_min4.csv")
```

```python
dirSpec = "/Users/alexgagliano/Documents/Research/2020oi/data/reference_data/"

spec = np.array([
'SN2013dx_SP1.csv',
'PTF10gvb.csv',
'sn2003aa-20030204-ui.flm',
'sn2007cl-20070524.470-ui.flm',
'sn2007gr-20070816.497-ui.flm'
])

spec_orig = np.array([
'sn2003aa-20030204-ui.flm',
'sn2007cl-20070524.470-ui.flm',
'sn2007gr-20070816.497-ui.flm'
])
```

```python
sns.set_context("poster")
for i in np.arange(len(spec)):
    s = spec[i]
    temp = pd.read_csv(dirSpec + s, header=None, names=['wavelength', 'flux'], delim_whitespace=True)
    if np.sum(temp['flux'] != temp['flux']) > 0:
        temp = pd.read_csv(dirSpec + s, header=None, names=['wavelength', 'flux'])
    plt.figure(figsize=(14,10))
    plt.plot(temp['wavelength'], temp['flux']/np.nanmean(temp['flux']), label=s)
    plt.plot(SN2020oi['wavelength(A)'], SN2020oi['flux(erg/scm2A)']/np.nanmean(SN2020oi['flux(erg/scm2A)']), c='k', alpha=0.7, label='SN 2020oi, -9 days')
    plt.xlabel(r"Wavelength (Angstroms)")
    plt.ylabel("Normalized Flux")
    plt.legend()
    plt.savefig("/Users/alexgagliano/Documents/Research/2020oi/img/2020oi_EarlySpectra_%s.png"%s.strip(".flm").strip(".csv"))
```

```python
plt.figure(figsize=(20,14))
#plt.plot(SN1983V['wavelength'], SN1983V['flux']/np.nanmean(SN1983V['flux']), alpha=0.5, ls='--')
plt.plot(PTF10gvb['wavelength'], PTF10gvb['flux']/np.nanmean(PTF10gvb['flux']), alpha=0.5, ls='--', label='iPTF10gvb, -13 days')
plt.plot(SN2020oi['wavelength(A)'], SN2020oi['flux(erg/scm2A)']/np.nanmean(SN2020oi['flux(erg/scm2A)']), label='SN 2020oi, -9 days', c='k')
plt.plot(SN2013dx_min8['wavelength'], SN2013dx_min8['flux']/np.nanmean(SN2013dx_min8['flux']), alpha=0.5, label='2013dx, -8 days')
for s in spec_orig:
    temp = pd.read_csv(dirSpec + s, header=None, names=['wavelength', 'flux'], delim_whitespace=True)
#    plt.figure(figsize=(10,7))
    plt.plot(temp['wavelength'], temp['flux']/np.nanmean(temp['flux']), label=s, alpha=0.5)
plt.legend()
plt.show()
```
