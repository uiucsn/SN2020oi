```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
 
# Create data
X = np.arange(0, 10, 1)
Y = X + 5 * np.random.random((5, X.size))
 
# There are 4 types of baseline we can use:
baseline = ["zero", "sym", "wiggle", "weighted_wiggle"]
 
# Let's make 4 plots, 1 for each baseline
for n, v in enumerate(baseline):
    if n<3 :
        plt.tick_params(labelbottom='off')
    plt.subplot(2 ,2, n + 1)
    plt.stackplot(X, *Y, baseline=v)
    plt.title(v)
    plt.tight_layout()
```

```python
Abundances = pd.read_csv("/Users/alexgagliano/Documents/Research/2020oi/data/derived_data/finalComposition.csv", delim_whitespace=True)
```

```python
elements = Abundances.columns[1:]
element_ab = []
for element in elements:
    element_ab.append(Abundances[element])
```

```python
sns.set_palette("muted")
sns.set_context("talk")
plt.figure(figsize=(10,10))
plt.stackplot(Abundances['Day'],element_ab,baseline='sym')
plt.xlim((11, 19))
plt.yscale("log")
plt.savefig("/Users/alexgagliano/Documents/Research/2020oi/img/finalComposition_stackPlot.png",dpi=200, bbox_inches='tight')
```
