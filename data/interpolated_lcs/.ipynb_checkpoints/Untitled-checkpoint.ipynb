```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
```

```python
SBO= pd.read_csv("/Users/alexgagliano/Documents/Outreach/ASB_Sonification/datasets/SN2005bf_photometry.csv")
```

```python
phot_sub = SBO[SBO['band'].isin(['u', 'g', 'B', 'V'])]
```

```python
phot_sub_B = phot_sub[phot_sub['band']=='B'].sort_values(by=['time']).drop_duplicates(subset=['time'])
phot_sub_V = phot_sub[phot_sub['band']=='V'].sort_values(by=['time']).drop_duplicates(subset=['time'])
phot_sub_u = phot_sub[phot_sub['band']=='u'].sort_values(by=['time']).drop_duplicates(subset=['time'])
phot_sub_g = phot_sub[phot_sub['band']=='g'].sort_values(by=['time']).drop_duplicates(subset=['time'])
```

```python
plt.plot(phot_sub_g['time'], phot_sub_g['magnitude'], 'o')
plt.axvline(x=53467)
plt.axvline(x=53530)
```

```python
xtest = phot_sub_B['time'].values
```

```python
from scipy.interpolate import interp1d

x = np.linspace(53467, 53530, num=100)

newB = np.interp(x, phot_sub_B['time'].values, phot_sub_B['magnitude'])
newV = np.interp(x, phot_sub_V['time'].values, phot_sub_V['magnitude'])
newu = np.interp(x, phot_sub_u['time'].values, phot_sub_u['magnitude'])
newg = np.interp(x, phot_sub_g['time'].values, phot_sub_g['magnitude'])

```

```python
sns.set_context("talk")
plt.figure(figsize=(10,7))
plt.plot(x, newB, 'o')
plt.plot(phot_sub_B['time'].values, phot_sub_B['magnitude'], 'o')
plt.gca().invert_yaxis()

plt.figure(figsize=(10,7))
plt.plot(x, newV, 'o')
plt.plot(phot_sub_V['time'].values, phot_sub_V['magnitude'], 'o')
plt.gca().invert_yaxis()

plt.figure(figsize=(10,7))
plt.plot(x, newu, 'o')
plt.plot(phot_sub_u['time'].values, phot_sub_u['magnitude'], 'o')
plt.gca().invert_yaxis()

plt.figure(figsize=(10,7))
plt.plot(x, newg, 'o')
plt.plot(phot_sub_g['time'].values, phot_sub_g['magnitude'], 'o')
plt.gca().invert_yaxis()
```

```python
SN2005bf = pd.DataFrame({'MJD':x, 'B':newB ,'V':newV, 'u':newu, 'g':newg})

SN2005bf['B_son'] = -SN2005bf['B'] + np.nanmax(SN2005bf['B'])
SN2005bf['V_son'] = -SN2005bf['V'] + np.nanmax(SN2005bf['V'])
SN2005bf['u_son'] = -SN2005bf['u'] + np.nanmax(SN2005bf['u'])
SN2005bf['g_son'] = -SN2005bf['g'] + np.nanmax(SN2005bf['g'])

SN2005bf.to_csv("/Users/alexgagliano/Documents/Outreach/ASB_Sonification/datasets/SN2005bf.csv",index=False)
```
